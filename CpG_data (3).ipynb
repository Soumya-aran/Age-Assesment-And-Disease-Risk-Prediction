KNN IMPLEMENTATION

```python


# Assuming you have loaded your dataset into a DataFrame named 'df'
# Let's say 'sample_type' is the target feature with 2 classes (binary classification)
# and 'disease' is the target feature with 10 classes (multiclass classification)

# Selecting features and target variables
X = data.drop(['sample_type', 'disease'], axis=1)  # Features
y_sample_type = data['sample_type']  # Target variable for sample type
y_disease = data['disease']  # Target variable for disease

# Splitting the dataset into training and testing sets for sample type
X_train_sample_type, X_test_sample_type, y_train_sample_type, y_test_sample_type = train_test_split(
    X, y_sample_type, test_size=0.2, random_state=42)

# Splitting the dataset into training and testing sets for disease
X_train_disease, X_test_disease, y_train_disease, y_test_disease = train_test_split(
    X, y_disease, test_size=0.2, random_state=42)


# Standardize features
scaler = StandardScaler()
X_train_sample_type_scaled = scaler.fit_transform(X_train_sample_type)
X_test_sample_type_scaled = scaler.transform(X_test_sample_type)

X_train_disease_scaled = scaler.fit_transform(X_train_disease)
X_test_disease_scaled = scaler.transform(X_test_disease)


# Create KNN classifier objects for both sample type and disease
knn_sample_type = KNeighborsClassifier(n_neighbors=5)
knn_disease = KNeighborsClassifier(n_neighbors=5)

# Training the KNN models for both sample type and disease
knn_sample_type.fit(X_train_sample_type, y_train_sample_type)
knn_disease.fit(X_train_disease, y_train_disease)

# Making predictions for both sample type and disease
y_pred_sample_type = knn_sample_type.predict(X_test_sample_type)
y_pred_disease = knn_disease.predict(X_test_disease)

# Confusion matrix and classification report for sample type
print("Confusion Matrix for Sample Type:")
print(confusion_matrix(y_test_sample_type, y_pred_sample_type))
print("\nClassification Report for Sample Type:")
print(classification_report(y_test_sample_type, y_pred_sample_type))

# Confusion matrix and classification report for disease
print("\nConfusion Matrix for Disease:")
print(confusion_matrix(y_test_disease, y_pred_disease))
print("\nClassification Report for Disease:")
print(classification_report(y_test_disease, y_pred_disease))
```

SVM IMPLEMENTATION

```python
from sklearn.svm import SVC




# Features and target variables
X = data.drop(['sample_type', 'disease'], axis=1)  # Assuming 'sample_type' and 'disease' are target features
y_sample_type = data['sample_type']
y_disease = data['disease']

# Split the dataset into train and test sets for sample type
X_train_sample_type, X_test_sample_type, y_train_sample_type, y_test_sample_type = train_test_split(X, y_sample_type, test_size=0.2, random_state=42)

# Split the dataset into train and test sets for disease
X_train_disease, X_test_disease, y_train_disease, y_test_disease = train_test_split(X, y_disease, test_size=0.2, random_state=42)

# Support Vector Machine for sample type
svm_sample_type = SVC(kernel='linear', decision_function_shape='ovo')  # One-vs-One strategy
svm_sample_type.fit(X_train_sample_type, y_train_sample_type)

# Support Vector Machine for disease
svm_disease = SVC(kernel='linear', decision_function_shape='ovo')  # One-vs-One strategy
svm_disease.fit(X_train_disease, y_train_disease)

# Predictions for sample type
y_pred_sample_type = svm_sample_type.predict(X_test_sample_type)

# Predictions for disease
y_pred_disease = svm_disease.predict(X_test_disease)

# Confusion matrix and classification report for sample type
print("Confusion Matrix for Sample Type:")
print(confusion_matrix(y_test_sample_type, y_pred_sample_type))
print("\nClassification Report for Sample Type:")
print(classification_report(y_test_sample_type, y_pred_sample_type))

# Confusion matrix and classification report for disease
print("\nConfusion Matrix for Disease:")
print(confusion_matrix(y_test_disease, y_pred_disease))
print("\nClassification Report for Disease:")
print(classification_report(y_test_disease, y_pred_disease))
```

MLP IMPLEMENTATION

```python
from sklearn.neural_network import MLPClassifier

features = data.drop(columns=['sample_type', 'disease'])
target_sample_type = data['sample_type']
target_disease = data['disease']

# Split the data into training and testing sets
X_train, X_test, y_train_sample_type, y_test_sample_type = train_test_split(features, target_sample_type, test_size=0.2, random_state=42)
X_train, X_test, y_train_disease, y_test_disease = train_test_split(features, target_disease, test_size=0.2, random_state=42)

# Define and train MLP model for 'sample type' classification
mlp_sample_type = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000)
mlp_sample_type.fit(X_train, y_train_sample_type)

# Define and train MLP model for 'disease' classification
mlp_disease = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000)
mlp_disease.fit(X_train, y_train_disease)

# Predict 'sample type' and 'disease' classes
y_pred_sample_type = mlp_sample_type.predict(X_test)
y_pred_disease = mlp_disease.predict(X_test)

# Print confusion matrix and classification report for 'sample type'
print("Confusion Matrix for Sample Type:")
print(confusion_matrix(y_test_sample_type, y_pred_sample_type))
print("\nClassification Report for Sample Type:")
print(classification_report(y_test_sample_type, y_pred_sample_type))

# Print confusion matrix and classification report for 'disease'
print("\nConfusion Matrix for Disease:")
print(confusion_matrix(y_test_disease, y_pred_disease))
print("\nClassification Report for Disease:")
print(classification_report(y_test_disease, y_pred_disease))
```

RF IMPLEMENTATION

```python
from sklearn.ensemble import RandomForestClassifier

X = data.drop(['sample_type', 'disease'], axis=1)
y_sample_type = data['sample_type']
y_disease = data['disease']

# Splitting data into train and test sets for each target feature
X_train, X_test, y_sample_type_train, y_sample_type_test = train_test_split(X, y_sample_type, test_size=0.2, random_state=42)
X_train, X_test, y_disease_train, y_disease_test = train_test_split(X, y_disease, test_size=0.2, random_state=42)

# Train Random Forest models for each target feature
rf_sample_type = RandomForestClassifier(n_estimators=100, random_state=42)
rf_disease = RandomForestClassifier(n_estimators=100, random_state=42)

rf_sample_type.fit(X_train, y_sample_type_train)
rf_disease.fit(X_train, y_disease_train)

# Predictions
y_sample_type_pred = rf_sample_type.predict(X_test)
y_disease_pred = rf_disease.predict(X_test)

# Evaluation: Confusion Matrix and Classification Report
print("Confusion Matrix and Classification Report for 'sample type':")
print(confusion_matrix(y_sample_type_test, y_sample_type_pred))
print(classification_report(y_sample_type_test, y_sample_type_pred))

print("\nConfusion Matrix and Classification Report for 'disease':")
print(confusion_matrix(y_disease_test, y_disease_pred))
print(classification_report(y_disease_test, y_disease_pred))
```

LOGISTIC REGRESSION IMPLEMENTATION


```python
from sklearn.linear_model import LogisticRegression


# Assuming df is your DataFrame with the dataset
# X contains the features and y contains the target variables
X = data.drop(['sample_type', 'disease'], axis=1)  # Features
y_sample_type = data['sample_type']  # Target variable for sample type
y_disease = data['disease']  # Target variable for disease

# Train-test split for sample type
X_train_sample_type, X_test_sample_type, y_train_sample_type, y_test_sample_type = train_test_split(X, y_sample_type, test_size=0.2, random_state=42)

# Train-test split for disease
X_train_disease, X_test_disease, y_train_disease, y_test_disease = train_test_split(X, y_disease, test_size=0.2, random_state=42)

# Logistic Regression for sample type
lr_sample_type = LogisticRegression(max_iter=1000)
lr_sample_type.fit(X_train_sample_type, y_train_sample_type)
y_pred_sample_type = lr_sample_type.predict(X_test_sample_type)

# Logistic Regression for disease
lr_disease = LogisticRegression(max_iter=1000)
lr_disease.fit(X_train_disease, y_train_disease)
y_pred_disease = lr_disease.predict(X_test_disease)

# Confusion matrix and classification report for sample type
print("Confusion Matrix for Sample Type:")
print(confusion_matrix(y_test_sample_type, y_pred_sample_type))
print("\nClassification Report for Sample Type:")
print(classification_report(y_test_sample_type, y_pred_sample_type))

# Confusion matrix and classification report for disease
print("\nConfusion Matrix for Disease:")
print(confusion_matrix(y_test_disease, y_pred_disease))
print("\nClassification Report for Disease:")
print(classification_report(y_test_disease, y_pred_disease))
```

DECISION TREE IMPLEMENTATION

```python
from sklearn.tree import DecisionTreeClassifier

# Assuming 'sample type' and 'disease' are your target features
X = data.drop(['sample_type', 'disease'], axis=1)  # Features
y_sample_type = data['sample_type']  # Target feature 1
y_disease = data['disease']  # Target feature 2

# Splitting the data into training and testing sets
X_train, X_test, y_sample_type_train, y_sample_type_test, y_disease_train, y_disease_test = train_test_split(
    X, y_sample_type, y_disease, test_size=0.2, random_state=42)

# Decision Tree model for sample type
sample_type_model = DecisionTreeClassifier()
sample_type_model.fit(X_train, y_sample_type_train)

# Decision Tree model for disease
disease_model = DecisionTreeClassifier()
disease_model.fit(X_train, y_disease_train)

# Predictions
sample_type_pred = sample_type_model.predict(X_test)
disease_pred = disease_model.predict(X_test)

# Confusion matrix and classification report for sample type
print("Confusion Matrix and Classification Report for Sample Type:")
print(confusion_matrix(y_sample_type_test, sample_type_pred))
print(classification_report(y_sample_type_test, sample_type_pred))

# Confusion matrix and classification report for disease
print("\nConfusion Matrix and Classification Report for Disease:")
print(confusion_matrix(y_disease_test, disease_pred))
print(classification_report(y_disease_test, disease_pred))
```
